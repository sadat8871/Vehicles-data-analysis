```python
# import python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# import csv file by python read_statement
vehicles=pd.read_csv(r"C:\Users\Sadat\Desktop\data\data\vehicles_dataset.csv")
```

```python
# top file starting data
vehicles.head(5)
```

```python
#top 5 lower data
vehicles.tail(5)
```

```python
# find the tota columns header 
vehicles.columns
```

```python
# Checking the null values 
vehicles.isnull().sum()
```

```python
# fill null values 
vehicle=vehicles.fillna({"description":"Na","engine":0,"price":0,"cylinders":"Na","fuel":"not available","mileage":0,"transmission":"NA","trim":"bfill",
                "body":"fill","doors":"nan", "exterior_color":"na","interior_color":"na",})

# note "bfill=beforefill,fill= after fill"
```

```python
# find new null values
vehicle.isnull().sum()
```

```python
# Distribution of vehicle prices with frequency
plt.figure(figsize=(10, 5))
sns.histplot(vehicle['price'], kde=True)
plt.title('Distribution of Vehicle Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Highest-selling vehicle customer like body type
plt.figure(figsize=(10, 5))
sns.histplot(vehicle['body'], kde=True)
plt.xticks(rotation='vertical')
plt.title('Customer like body type')
plt.xlabel('body')
plt.ylabel('Frequency')
plt.show()
```

```python
# Highest-selling vehicle by maker
plt.figure(figsize=(12, 6))
sns.countplot(data=vehicle, x='make', order=vehicle['make'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Vehicle Count by Make')
plt.xlabel('Make')
plt.ylabel('Count')
plt.show()
```

```python
# Categorical variable analysis - Type
plt.figure(figsize=(12, 6))
sns.countplot(data=vehicle, x='type', order=vehicle['type'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Vehicle Count by Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Average price by make
avg_price_by_make = vehicle.groupby('make')['price'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
avg_price_by_make.plot(kind='bar')
plt.title('Average Price by Make')
plt.xlabel('Make')
plt.ylabel('Average Price')
plt.show()
```
